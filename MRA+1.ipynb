```python
import matplotlib.pyplot as plt
import matplotlib.style
plt.style.use('classic')
%matplotlib inline 
#importing seaborn for statistical plots
import seaborn as sns
sns.set(context="notebook", palette="Spectral", style = 'darkgrid',
       font_scale = 1, color_codes=True)
```

```python
fig, axes = plt.subplots(nrows=4,ncols=2)
fig.set_size_inches(20, 20)

a = sns.distplot(Sales_df['QUANTITYORDERED'] , ax=axes[0][0])
a = sns.boxplot(Sales_df['QUANTITYORDERED'] , ax=axes[0][1])
a = sns.distplot(Sales_df['PRICEEACH'] , ax=axes[1][0])
a = sns.boxplot(Sales_df['PRICEEACH'] , ax=axes[1][1])
a = sns.distplot(Sales_df['MSRP'] , ax=axes[2][0])
a = sns.boxplot(Sales_df['MSRP'] , ax=axes[2][1])
a = sns.distplot(Sales_df['SALES'] , ax=axes[3][0])
a = sns.boxplot(Sales_df['SALES'] , ax=axes[3][1])
plt.tight_layout()
plt.show()
```

```python
corr = Sales_df.corr()
print(corr)
```
